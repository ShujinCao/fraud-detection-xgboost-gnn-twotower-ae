```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("../data/processed/sim_gnn_clusters.csv")
df["cluster"] = df["cluster"].astype(str)

normal = df[df["outlier"] == 0]
outliers = df[df["outlier"] == 1]

fig = px.scatter(
    normal,
    x="x",
    y="y",
    color="cluster",
    title="GNN Clusters (DBSCAN Outliers Highlighted)",
    hover_data=["fraud_score", "claim_id"],
)

fig.add_scatter(
    x=outliers["x"],
    y=outliers["y"],
    mode="markers",
    marker=dict(symbol="x", size=9, color="black"),
    name="Outliers",
)

fig.show()
```
